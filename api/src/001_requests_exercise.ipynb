```python
# Load Env Variables
from dotenv import load_dotenv
load_dotenv()

# Create API Client
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5" # fastest model, but also least capable. For more capable models, see https://docs.anthropic.com/claude/reference/models

# Create Helper functions
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python

# make a list of messages
messages = []

# Use a 'While true' loop to run the chatbot forever
while True:
    user_input = input("> ")
    add_user_message(messages, user_input)
    answer = chat(messages)
    add_assistant_message(messages, answer)
    print(answer)
```
