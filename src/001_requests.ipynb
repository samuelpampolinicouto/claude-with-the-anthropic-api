```python
# Load Env Variables
from dotenv import load_dotenv
load_dotenv()
```

```python
# Create API Client
from anthropic import Anthropic
client = Anthropic()
model = "claude-haiku-4-5" # fastest model, but also least capable. For more capable models, see https://docs.anthropic.com/claude/reference/models
```

```python
# Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {"role": "user", "content": "What is quantum computing? Answer in one sentence."}
    ]
)
```
